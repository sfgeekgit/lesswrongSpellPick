```python
import pandas as pd
import torch
import torch.nn as nn
import fastai
from fastai.tabular.all import *
```

```python
# Load the CSV file
df = pd.read_csv('dnd_mag.csv')
df.head()
```

```python
df[:10]
```

```python
# Define your target variable and continuous feature names
y_names = "Mage A Wins?"
cont_names = ['Fire', 'Water', 'Earth', 'Air', 'Light']

# Assuming all other columns are categorical (including boolean features)
cat_names = [col for col in df.columns if col not in cont_names + [y_names]]

# Step 2: Prepare your data
# dls = TabularDataLoaders.from_csv('dnd_mag.csv', y_names=y_names,
dls = TabularDataLoaders.from_df(df, y_names=y_names,
                                   y_block=CategoryBlock,
                                   cont_names=cont_names, cat_names=cat_names,
                                   procs=[Categorify, FillMissing, Normalize])

dls.show_batch()
```

```python


# Learner
learn = tabular_learner(dls, metrics=accuracy)
```

```python
import itertools

test_row = df.iloc[0].copy()

test_row['Fire'] = 24
test_row['Water'] = 39
test_row['Earth'] = 26
test_row['Air'] = 26
test_row['Light'] = 18

for name in cat_names:
  test_row[name] = 0

test_row["Mage B Has Flambeau's Flying Fireball"] = 1
test_row["Mage B Has Merinita's Mud Missiles"] = 1
test_row["Mage B Has Solomon's Solar Shield"] = 1


indexes = [i for i in range(5,17)]
combos = itertools.combinations(indexes,3)
all_rows = []

for index_set in combos:
  row = test_row.copy()
  for i in index_set:
    row[i] = 1
  all_rows.append(row.copy())

test_df = pd.DataFrame(all_rows)
test_dl = learn.dls.test_dl(test_df)

preds = learn.get_preds(dl=test_dl)
max = 146 #preds[0][:,1].topk(3)
print(max)
print(preds[0][max,1])

row, clas, probs = learn.predict(all_rows[max])
print(probs)

print(all_rows[max])
```
